<span 
style="color:grey">En python todo es un objeto que tiene propiedades y métodos.</span>

<p style='text-align: justify;'>Sí, en Python, todo es un objeto y tiene atributos y métodos. Los atributos son las características o variables de 
un objeto, mientras que los métodos son las acciones o funciones que puede realizar el objeto. Cada objeto en Python pertenece a una clase, que 
define su comportamiento y estado.</p>

<span 
style="color:grey">Las funciones son fragmentos de codigo reutilizable.</span>

<p style='text-align: justify;'>Sí, las funciones son fragmentos de código que se pueden reutilizar varias veces en un programa. Permiten 
encapsular lógica específica en un bloque de código independiente, lo que hace que el código sea más organizado, fácil de entender y mantener. 
Además, al utilizar funciones, se pueden evitar repeticiones de código y reducir la complejidad del programa. Las funciones también se pueden 
pasar como argumentos a otras funciones y pueden devolver valores o resultados para ser utilizados en otras partes del programa.</p>

<span 
style="color:grey">Las clases son fabricas de objetos que tienen similares atributos y propiedades.</span>

<p style='text-align: justify;'>Sí, las clases son un mecanismo de programación orientada a objetos (POO) que se utilizan para definir un conjunto 
de atributos y métodos (funciones) que son comunes a un conjunto de objetos. Una clase es como una plantilla o molde a partir del cual se pueden 
crear objetos o instancias de esa clase, cada uno con su propio conjunto de atributos y métodos.</p>

<p style='text-align: justify;'>Los atributos son las variables que contienen los datos o estado de un objeto, mientras que los métodos son las 
funciones que definen las acciones o comportamiento que puede realizar un objeto.
Al utilizar clases se pueden agrupar los objetos en categorías lógicas, lo que hace que el código sea más fácil de entender y mantener. Además, la
reutilización de código es mucho más fácil al crear objetos a partir de una clase existente en lugar de escribir código desde cero cada vez.</p>

*Ejemplo de operaciones aritmeticas:*

```python
print(2 + 3)   # suma(+)
print(3 - 1)   # resta(-)
print(2 * 3)   # multipicacion(*)
print(3 / 2)   # division(/)
print(3 ** 2)  # exponencial(**)
print(3 % 2)   # modulo(%)
print(3 // 2)  # Operador division de piso(//)
```

<span style="font-size:1.5em;"><span style="color:grey">Chequeando tipos de datos</span></span>

En Python, la función `type()` es una función incorporada que se utiliza para determinar el tipo de un `objeto`. Toma un objeto como argumento y 
devuelve una referencia al objeto que representa el tipo del argumento. La función `type()`` es útil para verificar el tipo de un objeto antes de 
realizar una operación en él, o para fines de depuración.

*Ejemplo:*

```python
print(type(10))                    # Int (entero)
print(type(3.14))                  # Float (decimal)
print(type(1 + 3j))                # Complex (complejo)
print(type('Asabeneh'))            # String (cadena de texto)
print(type([1, 2, 3]))             # List (lista)
print(type({'name': 'Asabeneh'}))  # Dictionary (diccionario)
print(type({9.8, 3.14, 2.7}))      # Set (conjunto)
print(type((9.8, 3.14, 2.7)))      # Tuple (tupla)
```

*Explicación de algunos términos:*

*`Objeto`*: Un objeto es una entidad en Python que tiene un estado y un comportamiento. Los objetos se crean a partir de clases.

*`Tipo`*: El tipo de un objeto determina sus propiedades y comportamientos.

*`Argumento`*: Un argumento es un valor que se pasa a una función.

*`Referencia`*: Una referencia es una forma de acceder a un objeto.

*`Clase`*: Una clase es un molde para crear objetos.

*`Instancia`*: Una instancia es un objeto creado a partir de una clase.

*`Parámetro`*: Un parámetro es una variable que se declara en la definición de una función.

*`Argumento por palabra clave`*: Un argumento por palabra clave es un argumento que se pasa a una función usando una palabra clave.

*`Metaclase`*: Una metaclase es una clase que se utiliza para crear otras clases.

*`Subclase`*: Una subclase es una clase que hereda de otra clase.
