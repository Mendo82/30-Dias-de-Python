## Resumen rápido de Selenium 4

Selenium 4, en general, genera mucha curiosidad ya que sigue una arquitectura diferente en comparación con su predecesor (es decir, 
Selenium 3). Antes de ver Python con Selenium, hagamos un resumen rápido de Selenium 4 (en su totalidad).

* Protocolo WebDriver W3C

    El protocolo JSON Wire utilizado para la comunicación entre las bibliotecas cliente de Selenium y el navegador web correspondiente se 
    retirará con Selenium 4.

    La introducción del protocolo WebDriver W3C es lo más destacado de Selenium 4. Una mayor estabilidad y menos descamación en las 
    pruebas de Selenium son las razones principales para cambiar a Selenium 4.

* Arquitectura Grid renovada

    Selenium Grid 4 admite una arquitectura mejorada con la inclusión de enrutador, distribuidor, mapa de sesión y nodo en un solo 
    archivo jar.

    Como el archivo Jar de Selenium Server también contiene todas las dependencias necesarias para ejecutar Grid, los usuarios de 
    Selenium Grid ya no tienen que iniciar Hub y Node por separado.

* Amplio conjunto de API de acción

    Las API de acción en Selenium 4 le permiten realizar acciones como acercar/alejar, acciones multitáctiles, pellizcar y acercar, y más.

    Con este nuevo conjunto de API de acción, los desarrolladores tienen la flexibilidad de realizar ciertas acciones en unos pocos (o un 
    solo) paso(s), que podrían haber tomado más pasos en versiones anteriores a Selenium 4.

* Selenium IDE 'muy útil'

    Ha habido mejoras significativas en Selenium IDE con Selenium 4. Tiene una interfaz de usuario mucho mejor. Además de Chrome, el IDE 
    también está disponible para Firefox y también se espera que llegue a Microsoft Edge.

    El corredor CLI llamado Selenium IDE Runner (o SIDE Runner) permite a los usuarios ejecutar pruebas grabadas entre navegadores en 
    paralelo. La introducción de mecanismos de flujo de control y la exportación de código con enlaces de idiomas oficiales como Java, 
    Python, Ruby, etc., son algunas de las otras características disponibles con el IDE de Selenium mejorado.

Ahora que hemos tocado los aspectos básicos de Selenium 4, profundicemos en Python con Selenium 4 y exploremos las funciones que se 
destacan en Python con Selenium 4 para desarrolladores.

### 'Localizadores relativos' para localizar elementos web.

Python en Selenium 4 o los enlaces del lenguaje Python para WebDriver admiten métodos de localizador relativo que se pueden usar con 
el with_tag_nameatributo. Los localizadores relativos también se conocen como 'localizadores amigables'.

La implementación del localizador relativo en Python indica las siguientes opciones de localizador relativo que ayudan a localizar 
elementos web que están cerca, encima, debajo, etc., de un elemento web específico.

```text
LOCALIZADOR RELATIVO         DESCRIPCIÓN

above                  WebElement que se va a ubicar está 'encima' de un elemento específico (o particular).
below                  WebElement a ubicar está 'debajo' de un elemento específico (o particular).
to_left_of             WebElement a ubicar está 'a la izquierda de' un elemento específico (o particular).
to_right_of            WebElement a ubicar está 'a la derecha de' un elemento específico (o particular).
near                   WebElement que se va a ubicar está "como máximo a 50 píxeles" del elemento especificado (o particular).
```

### Cómo usar el método find_elements para localizadores relativos en Selenium 4 Python.

El método find_elements en WebDriver tiene los cambios de implementación correspondientes necesarios para devolver localizadores 
relativos. Los cambios están marcados en color 'Azul':

<img src='Imagenes/imagen_4.png'>

Como se mencionó anteriormente, tenemos que usar el find_elementsmétodo en lugar del find_elementmétodo para usar localizadores 
relativos en Selenium 4 (con Python). Dado que el find_elementsmétodo devuelve una lista de elementos web que coinciden con la 
estrategia del localizador web dada, debemos analizar la lista para realizar acciones relevantes en el elemento web 'relativo' 
requerido.

Los métodos de localización relativa de Selenium 4 (es decir, findElements en Python) admiten el uso con el with_tag_nameatributo. 
Para usar localizadores relativos en Selenium 4 Python, realice los siguientes pasos:

1. Localice el elemento web 'relativo' (es decir, cerca, derecha, izquierda, arriba, abajo, etc.) al que se deben rastrear los 
elementos correspondientes. Utilice el find_elementmétodo con localizadores web como ID, XPATH, NAME, LINK_TEXT, CSS_SELECTOR, etc., 
para ubicar el elemento web.

1. Utilice el find_elementsmétodo que acepta el método with_tag_name(p. ej., nombre de la etiqueta: "entrada") junto con el
localizador relativo pertinente (p. ej., arriba/abajo/a_la_izquierda_de/a_la_derecha_de/cerca). El resultado es una lista de 
elementos web que coinciden con el criterio requerido.

1. Analice la lista de elementos web obtenida en el paso (b) y realice las acciones pertinentes en los elementos web requeridos.

La ventaja de ubicar elementos que están cerca de otros elementos con el find_elementmétodo es que puede realizar acciones relevantes
en muchos elementos web que coinciden con la estrategia del localizador relativo. Esto no hubiera sido posible con el 
find_elementmétodo capaz de devolver solo un único elemento web que coincidiera con la estrategia del localizador web.

### Cómo usar localizadores relativos en Python con Selenium 4

Con la versión Alpha actual, el find_elementmétodo no se puede usar con localizadores relativos para encontrar un WebElement 
específico. Ya sea que tenga que usar un único localizador relativo o varios localizadores relativos para ubicar "un" WebElement 
específico, aún necesita usar el find_elementsmétodo.

La ejecución de las pruebas que demuestran los localizadores relativos se ejecuta en Selenium Grid basado en la nube de LambdaTest. 
Las capacidades del navegador se generan mediante LambdaTest Capabilities Generator. Después de crear una cuenta en LambdaTest, se 
recomienda anotar el nombre de usuario y la clave de acceso disponibles en la sección Perfil de LambdaTest.

### Demostración 1: Python con localizadores relativos de Selenium 4

### Escenario de prueba

1. Vaya a https://4dvanceboy.github.io/lambdatest/lambdasampleapp.html.
2. Haga clic en los elementos li2, li3 y li4.
3. Agregue un nuevo elemento, 'Pruebas felices en LambdaTest', a la lista (usando to_right_of relativeel localizador).
4. Agregue otro elemento, 'Pruebas entre navegadores en LambdaTest', a la lista (usando el localizador relativo cercano).

*Implementación:*


````python
# https://github.com/SeleniumHQ/selenium/blob/trunk/py/selenium/webdriver/support/relative_locator.py
# https://github.com/SeleniumHQ/selenium/blob/4c5b92bac07b17e223917c31caddf7035c120ea7/py/selenium/webdriver/remote/webdriver.py

import pytest
from selenium import webdriver
import sys
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from time import sleep
import urllib3
import warnings
from selenium.webdriver.support.relative_locator import with_tag_name
import json

# Configuración de capacidades del navegador para LambdaTest
browser_capabilities = {
        "user" : "user.name",
        "accessKey" : "Access-Key",
        "build" : "[Python] - Relative Locators in Selenium 4 on LambdaTest Login Page",
        "name" : "[Python] - Relative Locators in Selenium 4 on LambdaTest Login Page",
        "platformName" : "OS X Mavericks",
        "browserName" : "Chrome",
        "browserVersion" : "67.0"
}

user_name = "user-name"
app_key = "access-key"


def test_lambdatest_todo_app():
    # Deshabilitar advertencias de solicitud no segura
    urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
    
    # URL remota para LambdaTest
    remote_url = "https://" + user_name + ":" + app_key + "@hub.lambdatest.com/wd/hub"
    web_driver = webdriver.Remote(command_executor = remote_url, desired_capabilities = browser_capabilities)
    
    # Abrir la página de inicio de sesión de LambdaTest
    web_driver.get('https://accounts.lambdatest.com/login')
    web_driver.maximize_window()
    
    # Localizar el elemento web li1
    elem_label = web_driver.find_element(By.CSS_SELECTOR,".orsignup")
    
    # Utilizar el nombre de la etiqueta y localizadores relativos para encontrar los elementos web entre ellos
    elementBox = web_driver.find_elements(with_tag_name("input").below(elem_label))
    
    for items in elementBox:
        print(items.get_attribute('name'))
        # Use the Attribute Name to get the details of the web element
        elem_name = items.get_attribute('name')
        
        # Verificar si el atributo 'name' es "email"
        if (elem_name == "email"):
            # Perform a click operation on the web element
            print("Inside Email Address")
            # Escribir la dirección de correo electrónico
            web_driver.find_element(By.NAME, elem_name).send_keys("Email@gmail.com")
        
        # Verificar si el atributo 'name' es "password"
        if (elem_name == "password"):
            # Perform a click operation on the web element
            print("Inside Password")
            # Escribir la contraseña
            web_driver.find_element(By.NAME, elem_name).send_keys("Himanshu1!")
            # Hacer clic en "Mostrar contraseña" para asegurarse de que se ingresó correctamente
            web_driver.find_element(By.XPATH, "//span[@class='input-group-text password-group-text']").click()
            # Salir del bucle para no procesar elementos después del botón "Iniciar sesión"
            break
    
    # No realizamos una operación de clic aquí. La misma operación también se puede realizar dentro del bucle for.
    # En ese caso, tendríamos que usar el atributo 'class' en lugar del atributo 'name'.
    sleep(2)
    
    # Hacer clic en el botón "Iniciar sesión" ya que hemos ingresado el nombre de usuario y la contraseña
    web_driver.find_element(By.CSS_SELECTOR, ".btn").click()
    sleep(2)
    
    # Verificar el título de la nueva ventana. Comprobar si los títulos coinciden o no
    window_title = web_driver.title
    print(window_title)
    expected_title = "Welcome"
    
    if expected_title in window_title:
        print("Login successful")
    else:
        print("Login not successful")
    print("Relative Locators in Selenium 4 on LambdaTest Login Page complete")
    
    # Liberar los recursos utilizados por el controlador de Selenium WebDriver
    web_driver.quit()
````

````python
# Soporte para Chrome DevTools
# https://www.selenium.dev/documentation/en/support_packages/chrome_devtools/
# Implementación de WebDriver para CDP en Selenium 4
# https://github.com/SeleniumHQ/selenium/blob/4c5b92bac07b17e223917c31caddf7035c120ea7/py/selenium/webdriver/chromium/webdriver.py#L133
# Emulación de geolocalización con Python en Selenium 4
# CDP solo está disponible para WebDriver local (en Chrome/Chromium)
# https://github.com/SeleniumHQ/selenium/blob/474d11671452ffc6830e3b9603d6e438c9cce8fd/py/selenium/webdriver/chromium/webdriver.py

import pytest
from selenium import webdriver
import sys
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from time import sleep
import urllib3
import warnings
from selenium.webdriver.support.relative_locator import with_tag_name
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
import json

def test_cdp_command():
    web_driver = webdriver.Chrome()
    web_driver.maximize_window()
    time.sleep(2)
    
    map_coord = {
        "latitude": 42.1408845,
        "longitude": -72.5033907,
        "accuracy": 100
    }
    
    web_driver.execute_cdp_cmd('Emulation.setGeolocationOverride', {
        "latitude": map_coord['latitude'],
        "longitude": map_coord['longitude'],
        "accuracy": map_coord['accuracy']
    })
    
    web_driver.get('https://locations.dennys.com/search.html/')
    time.sleep(2)
    
    location_icon = web_driver.find_element(By.CSS_SELECTOR, ".icon-geolocate")
    time.sleep(2)
    location_icon.click()
    time.sleep(5)
    
    # Liberar recursos retenidos por el controlador de Selenium WebDriver
    web_driver.quit()
    
    print("Prueba de geolocalización con Selenium está completa")
````

Python con Selenium 4 también le permite emular diferentes tipos de red (o tipos de conexión). Según el tipo de conexión (es decir,

La emulación de las condiciones de red mediante el método set_network_conditions en Python con Selenium 4 le permite configurar la

### Escenario de prueba

1. Abra el navegador Chrome en el modo sin cabeza. Vaya a https://www.lambdatest.com.

rendimiento de carga: (500 * 1024).

3. Imprima las condiciones de la red para verificar si se configuraron correctamente o no.

*Implementación*

````python
# Emulate Network Conditions
# https://github.com/SeleniumHQ/selenium/blob/474d11671452ffc6830e3b9603d6e438c9cce8fd/py/selenium/webdriver/chromium/webdriver.py

import pytest
from selenium import webdriver
import sys
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from time import sleep
import urllib3
import warnings
from selenium.webdriver.support.relative_locator import with_tag_name
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.chrome.options import Options
import json

def test_emulate_network_settings():
    chrome_options = Options()  
    chrome_options.add_argument("--headless") 
    web_driver = webdriver.Chrome(options=chrome_options)
    web_driver.maximize_window()
    time.sleep(2)
    
    web_driver.set_network_conditions(
        offline=False,
        latency=10,
        download_throughput=500 * 1024,
        upload_throughput=500 * 1024)
    
    web_driver.get('https://www.lambdatest.com')
    net_con = web_driver.get_network_conditions()
    print()
    print(net_con)
    time.sleep(5)
    print("Latency: " + str(net_con["latency"]))
    print("Download Throughput: " + str(net_con['download_throughput']))
    print("Upload Throughput: " + str(net_con['upload_throughput']))
    print()
    
    # Liberar recursos retenidos por el controlador de Selenium WebDriver
    web_driver.quit()
    print("Prueba de emulación de configuración de red está completa")
````

### Manipulación del navegador: Cambiar entre ventanas o pestañas

El manejo de múltiples ventanas del navegador en Selenium y pestañas es uno de los escenarios de prueba típicos para la prueba 
automatizada del navegador. Para cambiar a una nueva ventana/pestaña, debe usar el identificador de ventana 'correcto' para que la 
ventana/pestaña deseada del navegador se enfoque.

El proceso de cambiar a una nueva ventana/pestañas implica recorrer la lista de ventanas o pestañas que Selenium WebDriver puede ver, 
publicar las cuales se cambia a la ventana/pestaña requerida.

El tedioso proceso de cambiar a nuevas ventanas o pestañas en Selenium Python está listo para cambiar con Selenium 4. La API 
new_window en Selenium 4 Python ayuda a crear una nueva ventana o nueva pestaña. Publicar la creación de una nueva ventana/pestaña, 
el control cambia automáticamente a ella.

- Abra una nueva pestaña y cambie a la misma

```python
driver.switch_to.new_window('tab')
```

- Abra una nueva ventana y cambie a la misma

```python
driver.switch_to.new_window('window')
```

Veamos un escenario de prueba simple para demostrar el uso del método new_window:

*Escenario de prueba*

1. Abra el navegador Chrome y visite https://www.lambdatest.com
2. Crear una nueva pestaña
3. Cree una nueva ventana y establezca la URL en https://www.lambdatest.com/blog/
4. Imprima el número de manijas de ventana (es decir, debe ser 3 en total)

*Implementación*

````python
import pytest
from selenium import webdriver
import sys
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from time import sleep
import urllib3
import warnings
from selenium.webdriver.support.relative_locator import with_tag_name
import json

def test_new_window_tab():
    web_driver = webdriver.Chrome()
    web_driver.get('https://www.lambdatest.com')
    web_driver.maximize_window()

    # Abre una nueva pestaña y cambia a ella
    web_driver.switch_to.new_window('tab')
    time.sleep(5)

    # Abre una nueva ventana y cambia a ella
    web_driver.switch_to.new_window('window')
    time.sleep(5)

    # Accede a una URL en la nueva ventana
    web_driver.get('https://www.lambdatest.com/blog')
    time.sleep(10)

    print("Número de ventanas abiertas: " + str(len(web_driver.window_handles)))
    print("Selenium 4: Cambio a una nueva pestaña o ventana completado")

    # Libera los recursos retenidos por el controlador de Selenium WebDriver
    web_driver.quit()
````
