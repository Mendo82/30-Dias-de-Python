15. Crea un diagrama que describa la clase Motocicleta y los dos objetos instanciados.

````python
class Motocicleta:
    # Atributos de clase
    estado = 'Nueva'
    motor = False

    # Metodos
    def __init__(self, color, matricula, combustible_litros, numero_ruedas, marca, modelo,
                 fecha_fabricacion, velocidad_punta, peso, cantidad_maxima_combustible):
        self.color = color
        self.maticula = matricula
        self.combustible_litros = combustible_litros
        self.numero_ruedas = numero_ruedas
        self.marca = marca
        self.modelo = modelo
        self.fecha_fabricacion = fecha_fabricacion
        self.velocidad_punta = velocidad_punta
        self.peso = peso
        self.cantidad_maxima_combustible = cantidad_maxima_combustible

    def informacion(self):
        print(f"""
              Matricula: {self.maticula}
              Marca: {self.marca}
              Modelo: {self.modelo}
              Fecha de fabricacion: {self.fecha_fabricacion}
              Color: {self.color}
              Peso: {self.peso}
              Velocidad punta: {self.velocidad_punta}
              """)

    def arrancar(self):
        if self.motor:
            print('El motor ya estaba arrancado.')
        else:
            self.motor = True
            print('Motor arrancado.')

    def detener(self):
        if not self.motor:
            print('No puedes parar el motor, porque ya está apagado. ¿No lo oyes?», dado que el motor por defecto está apagado.')
        else:
            self.motor = False
            print('El motor se ha parado.')

    def consulta_precio(self):
        print(f'El precio de la motocicleta {self.marca} {self.modelo}, es de {self.precio}')

    def comprobar_combustible(self):
        print(f'---->Reporte del deposito de {self.marca} {self.modelo}<----')
        print(f'El deposito tiene {self.combustible_litros} litros.')
        self.cantidad_para_llenado_deposito = self.cantidad_maxima_combustible - self.combustible_litros
        print(f'Necesitas {self.cantidad_para_llenado_deposito} litros para llenar el deposito.\n')

    def repostar(self):
        lleno = True
        while lleno:
            self.llenado_combustible = float(
                input(f'Introduce la cantidad de combustible que quieres '
                      f'repostar para tu {self.marca}: ')
            )
            if self.llenado_combustible + self.combustible_litros > self.cantidad_maxima_combustible:
                print(
                    'No puede rebasar la capacidad maxima de combustible del deposito.'
                )
            else:
                self.combustible_litros += self.llenado_combustible
                print(
                    f'El deposito de combustible de tu {self.marca} {self.modelo} tiene ahora {self.combustible_litros} litros'
                )
                lleno = False


# Instancias de la clase Motocicleta
motocicleta_yamaha_1 = Motocicleta(
    "Roja y blanca",
    "45663-FHDY",
    10,
    2,
    "Yamaha",
    "YZF-R1",
    "20/02/2020",
    288,
    199,
    17
)

motocicleta_harley_1 = Motocicleta(
    matricula="4865-FHZ",
    combustible_litros=0,
    color="Negra",
    marca="Harley Davidson",
    modelo="Fat Boy",
    numero_ruedas=2,
    peso=304,
    fecha_fabricacion="29/09/2020",
    velocidad_punta=160,
    cantidad_maxima_combustible=20
)

motocicleta_harley_1.informacion()
motocicleta_harley_1.detener()
motocicleta_harley_1.arrancar()

motocicleta_harley_1.precio = 2700
print(
    f'El precio de la motocicleta {motocicleta_harley_1.marca} modelo {motocicleta_harley_1.modelo} es de {motocicleta_harley_1.precio}'
)
motocicleta_harley_1.consulta_precio()
# motocicleta_yamaha_1.consulta_precio() # AttributeError: ‘Motocicleta’ object has no attribute ‘precio’
motocicleta_harley_1.comprobar_combustible()
motocicleta_yamaha_1.comprobar_combustible()
motocicleta_harley_1.repostar()
motocicleta_yamaha_1.repostar()
````
