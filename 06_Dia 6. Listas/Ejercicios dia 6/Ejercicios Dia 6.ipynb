Declare una variable de lista llamada it_companies y asigne valores iniciales Facebook, Google, Microsoft, Apple, IBM, Oracle y Amazon.

La siguiente es una lista de las edades de 10 estudiantes:

```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
```

Ordena la lista y encuentra la edad mínima y máxima.

````python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]

ages.sort()
print(ages)
print('La edad minima es:', min(ages))
print('La edad maxima es:', max(ages))

# Agregue la edad mínima y la edad máxima nuevamente a la lista.

edad_minima_maxima = [min(ages), max(ages)]
ages.extend(edad_minima_maxima)
print(ages)
````

---

Encuentre la edad promedio (suma de todos los elementos dividida por su número).

````python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
suma = sum(ages)
num_items = len(ages)
print('La edad media es de:', suma / num_items)
````

---

Encuentre el rango de las edades (max menos min)

````python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
ages.sort()
indice_menor = ages.index(min(ages))
indice_mayor = ages.index(max(ages))
print(ages[indice_menor+1:indice_mayor])
````
